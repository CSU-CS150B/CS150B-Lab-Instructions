````python
import csv

def csv_data(filename):
    """
    Reads a CSV file and returns its content as a list of lists, skipping the header.
    """
    data = []
    with open(filename, 'r') as myfile:
        file_list = csv.reader(myfile)
        next(file_list) # Skip the header row
        for row in file_list:
            data.append(row)
    return data

def get_filtered_CSV(filename, filter_by):
    """
    Filters CSV data by a specified genre (case-insensitive) and returns a list of matching rows.
    """
    filtered_list = []
    all_data = csv_data(filename)
    for row in all_data:
        # Assuming genre is at index 3 and release year at index 2
        if len(row) > 3 and row[3].lower() == filter_by.lower():
            filtered_list.append(row)
    return filtered_list

def find_era(filename, genre, earliest, latest):
    """
    Finds songs of a specific genre within a given release year range.
    """
    find_songs = []
    song_genre = get_filtered_CSV(filename, genre)

    # Convert earliest and latest to integers
    earliest_year = int(earliest)
    latest_year = int(latest)

    for row in song_genre:
        # Ensure the row has enough elements before accessing index 2
        if len(row) > 2:
            try:
                year = int(row[2]) # Convert release year to integer

                if earliest_year <= year <= latest_year:
                     find_songs.append(row)
            except ValueError:
                 # Handle cases where the year might not be a valid integer
                 print(f"Skipping row with invalid year data: {row}")

    return find_songs

# Example Test calls (requires 'songs.csv' from previous steps):
# Create a dummy songs.csv for testing if it doesn't exist
# For local testing, you can create a file named 'songs.csv' with content like:
# Artist,Song Name,Release Year,Genre
# The Beatles,Let it Be,1970,Rock
# ACDC,Highway to Hell,1979,Rock
# Britney Spears,Toxic,2003,Pop
# Ed Sheeran,Shape of You,2017,Pop
# Queen,Bohemian Rhapsody,1975,Rock
# Michael Jackson,Thriller,1982,Pop

# Test with a dummy file if needed
# with open('songs.csv', 'w') as f:
#     f.write('Artist,Song Name,Release Year,Genre\n')
#     f.write('The Beatles,Let it Be,1970,Rock\n')
#     f.write('ACDC,Highway to Hell,1979,Rock\n')
#     f.write('Britney Spears,Toxic,2003,Pop\n')
#     f.write('Ed Sheeran,Shape of You,2017,Pop\n')
#     f.write('Queen,Bohemian Rhapsody,1975,Rock\n')
#     f.write('Michael Jackson,Thriller,1982,Pop\n')

print('Testing find_era function:')

rock_70s = find_era("songs.csv", "rock", "1970", "1980")
print("\nRock Songs from the 70s:")
if rock_70s:
    for song in rock_70s:
        print(song)
else:
    print("No Rock songs found from 1970-1980.")

pop_2000s = find_era("songs.csv", "pop", "2000", "2010")
print("\nPop Songs from the 2000s:")
if pop_2000s:
    for song in pop_2000s:
        print(song)
else:
    print("No Pop songs found from 2000-2010.")

print("\nTesting with a genre that might not exist or empty range:")
hiphop_90s = find_era("songs.csv", "hiphop", "1990", "1999")
print("Hip-Hop Songs from the 90s:")
if hiphop_90s:
    for song in hiphop_90s:
        print(song)
else:
    print("No Hip-Hop songs found from 1990-1999 (expected for dummy data).")
````
